```python
# Análisis de retención
retention_analysis = (
    customer_segmentation_final
    .groupBy("is_active")
    .agg(
        count("IDCLIENTE").alias("customer_count"),
        spark_sum("monetary_value").alias("total_revenue")
    )
    .withColumn(
        "status",
        when(col("is_active") == True, "Active (last 90 days)")
        .otherwise("Inactive (>90 days)")
    )
    .select("status", "customer_count", "total_revenue")
)
```
